```python
from ultralytics import FastSAM
from ultralytics import SAM
from PIL import Image
```

# 二：单张图片实例分割预测

```python
%%time
# Define an inference source
source = '/content/dataset/dog2.jpg'

# Load a model
model = SAM('sam_b.pt')

# Display model information (optional)
model.info()

# Run inference
result = model(source,  imgsz=1024, conf=0.95, iou=0.9)
```

```python
one_result = result[0].plot()

Image.fromarray(one_result)
```
